## 1.1 Why Python?

Python is a high-level, general-purpose programming language that emphasizes readability, conciseness, and developer productivity. It was created by Guido van Rossum and first released in 1991.

For students with a C++ and Java background, Python represents a shift in design philosophy:
- **C++** prioritizes low-level control and performance.
- **Java** emphasizes type safety and portability through the JVM.
- **Python** optimizes for rapid iteration and expressive syntax.

The trade-off is reduced raw execution speed, which is acceptable for most engineering workflows because the bottleneck is typically I/O, not computation.

---
## 1.2 Interpreted vs. Compiled Execution

| Language | Workflow |
|----------|----------|
| **C++**  | Source → Compiler → Machine code → Execute |
| **Java** | Source → javac → Bytecode (.class) → JVM |
| **Python (CPython)** | Source → Interpreter → Bytecode (in memory / .pyc) → CPython VM |

**Key Consequence:** No separate compilation step. Errors are detected at runtime, not at compile time.

> Because Python is dynamically typed and interpreted, type errors that C++ or Java would catch at compile time only surface when the offending line is executed. This makes thorough testing especially important.

## 1.5 The `__name__` Variable and the `__main__` Guard

Every Python file (module) has a built-in variable called `__name__`:
- If the file is **run directly** → `__name__` is set to `"__main__"`
- If the file is **imported** → `__name__` is set to the module's name

````python
# In a notebook or script run directly, __name__ is always "__main__"
print(f"__name__ = {__name__}")
````

### The Problem: Unwanted Code Execution on Import

Without a guard, test/demo code at the top level runs when the module is imported:

````python
# math_utils.py — WITHOUT __name__ guard (BAD)

def add(a, b):
    return a + b

def multiply(a, b):
    return a * b

# This code runs EVERY time the module is loaded — even on import!
print("Testing math_utils...")
print(f"add(2, 3) = {add(2, 3)}")
print(f"multiply(4, 5) = {multiply(4, 5)}")
print("All tests passed!")
````

### The Solution: `if __name__ == "__main__"`

````python
# math_utils.py — WITH __name__ guard (GOOD)

def add(a, b):
    return a + b

def multiply(a, b):
    return a * b

if __name__ == "__main__":
    # This block ONLY runs when you do: python math_utils.py
    # It does NOT run when another file does: import math_utils
    print("Testing math_utils...")
    print(f"add(2, 3) = {add(2, 3)}")
    print(f"multiply(4, 5) = {multiply(4, 5)}")
    print("All tests passed!")
````

> **Best Practice:** Always use the `if __name__ == "__main__"` guard in any Python file that could be both run directly and imported.

## 1.7 Virtual Environments and Package Management

A virtual environment is an isolated directory tree that contains a specific Python interpreter and a set of installed packages. This prevents version conflicts between projects.

### Using `venv` (built-in)
```bash
# Create a virtual environment
python -m venv myenv

# Activate (Windows)
myenv\Scripts\activate

# Activate (Linux / macOS)
source myenv/bin/activate

# Install packages
pip install requests numpy

# Freeze current dependencies
pip freeze > requirements.txt

# Reproduce environment elsewhere
pip install -r requirements.txt
```

### Using `conda` (recommended for this course)
```bash
# Create a new environment
conda create --name myproject python=3.12

# Activate the environment
conda activate myproject

# Install packages
conda install numpy pandas matplotlib jupyter

# Export environment to a file
conda env export > environment.yml

# Recreate from file
conda env create -f environment.yml
```

> Always work inside a virtual environment to keep your system Python clean and ensure reproducibility.

## 1.8 Development Workflow Overview

A typical Python development cycle:

1. **Explore** — Use the REPL to test small ideas and examine APIs.
2. **Write** — Organize code into modules and packages within a project directory.
3. **Test** — Run unit tests with `pytest` or `unittest` after each change.
4. **Debug** — Use `print()`, the built-in `pdb` debugger, or IDE breakpoints.
5. **Ship** — Package the project and manage dependencies with `pip` and `pyproject.toml`.

This tight feedback loop is one of Python's greatest strengths. Unlike C++ or Java, there is no waiting for a compilation step.
