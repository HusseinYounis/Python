## 2.1 Names and Objects

In Python, a variable is a **name** that references an **object** in memory. The type belongs to the object, not the name. This is called **name binding**.

```python
# Name binding — x can refer to different types
x = 42
print(f"x = {x}, type = {type(x)}")

x = "hello"
print(f"x = {x}, type = {type(x)}")
```

### Identity, Type, and Value

Every object has three properties:
- **Identity** — unique integer from `id(obj)`
- **Type** — from `type(obj)`
- **Value** — the data it contains

```python
a = 256
b = 256
print(f"a is b: {a is b}")   # True — CPython caches small integers [-5, 256]

c = 1000
d = 1000
print(f"c is d: {c is d}")   # May be False — depends on implementation
print(f"c == d: {c == d}")   # True — values are equal
```

> Use `==` to compare **values** and `is` to compare **identity** (same object in memory).

## 2.2 Numeric Types

| Type | Description | C++ Equivalent |
|------|-------------|----------------|
| `int` | Arbitrary-precision integers | No direct equivalent (unlimited!) |
| `float` | IEEE 754 double-precision | `double` |
| `complex` | Complex numbers (`3 + 4j`) | `std::complex<double>` |

```python
# Arbitrary-precision integers — no overflow!
big = 2 ** 200
print(f"2^200 = {big}")
print(f"Type: {type(big)}")
print(f"Number of digits: {len(str(big))}")
```

```python
import sys

# Memory growth of Python integers
for exp in [10, 100, 1000, 10_000, 100_000]:
    n = 2 ** exp
    size = sys.getsizeof(n)
    digits = len(str(n))
    print(f"2^{exp:>7,}  →  {digits:>6,} digits  →  {size:>8,} bytes")
```

### Arithmetic Operators

```python
a, b = 17, 5

print(f"{a} + {b}  = {a + b}")     # Addition
print(f"{a} - {b}  = {a - b}")     # Subtraction
print(f"{a} * {b}  = {a * b}")     # Multiplication
print(f"{a} / {b}  = {a / b}")     # True division (always float)
print(f"{a} // {b} = {a // b}")    # Floor division
print(f"{a} % {b}  = {a % b}")     # Modulo
print(f"{a} ** {b} = {a ** b}")    # Exponentiation
```

> **Key difference:** In C++/Java, `17 / 5` gives `3` (integer division). In Python, `/` always gives true division. Use `//` for floor division.

### Readable Numeric Literals — Underscore Separators

```python
# Underscores make large numbers readable (ignored by interpreter)
population = 8_100_000_000
budget = 4_500_000.75

print(population)
print(budget)

# Works with any numeric type
binary  = 0b1010_0001_1100_0011
hex_val = 0xFF_EC_DA
print(f"binary:  {binary}")
print(f"hex_val: {hex_val}")
```

### Number Bases: Binary, Octal, Decimal, Hexadecimal

```python
# Writing integers in different bases
decimal = 255
binary  = 0b11111111
octal   = 0o377
hexa    = 0xFF

print(f"All equal? {decimal == binary == octal == hexa}")

# Converting between bases
n = 255
print(f"bin: {bin(n)}")
print(f"oct: {oct(n)}")
print(f"hex: {hex(n)}")

# Parsing strings to int
print(f"int('FF', 16) = {int('FF', 16)}")
print(f"int('11111111', 2) = {int('11111111', 2)}")
```

```python
# Formatted output in different bases
for i in [8, 16, 32, 64, 128, 255]:
    print(f"{i:>4d}  {i:#06x}  {i:#010b}")
```

## 2.3 Strings

Strings are **immutable** sequences of Unicode characters. Python has no separate `char` type.

```python
name = "Ada Lovelace"
print(f"First char: {name[0]}")
print(f"Length: {len(name)}")
print(f"Upper: {name.upper()}")
print(f"Split: {name.split()}")
print(f"Joined: {'-'.join(name.split())}")
```

### Indexing and Slicing

```python
text = "PYTHON"
#        P  Y  T  H  O  N
#        0  1  2  3  4  5    ← positive index
#       -6 -5 -4 -3 -2 -1   ← negative index

print(f"text[0]   = {text[0]}")
print(f"text[-1]  = {text[-1]}")
print(f"text[0:3] = {text[0:3]}")
print(f"text[3:]  = {text[3:]}")
print(f"text[::-1]= {text[::-1]}")
```

```python
# Slicing with step
text = "PYTHON"
print(f"Every 2nd:    {text[::2]}")
print(f"Reversed:     {text[::-1]}")
print(f"Last 3:       {text[-3:]}")
print(f"All but last: {text[:-1]}")
```

### String Formatting (f-strings)

```python
voltage = 3.3
unit = "V"
print(f"Output: {voltage:.2f} {unit}")

name = "Alice"
age = 25
print(f"{name} is {age} years old.")
```

## 2.4 Booleans and Comparison Operators

```python
# Comparison operators
print(f"5 > 3:   {5 > 3}")
print(f"5 == 5.0: {5 == 5.0}")

# Boolean operators (words, not symbols)
print(f"True and False: {True and False}")
print(f"True or False:  {True or False}")
print(f"not True:       {not True}")

# Chained comparisons (unique to Python!)
x = 7
print(f"1 < {x} < 10: {1 < x < 10}")
```

### Short-Circuit Evaluation

Python's `and`/`or` return the **actual value** that decided the result, not just `True`/`False`.

```python
# `and` — short-circuits on first falsy
print(f"0 and 42:       {0 and 42}")
print(f"1 and 42:       {1 and 42}")

# `or` — short-circuits on first truthy
print(f"42 or 0:        {42 or 0}")
print(f"'' or 'default': {'' or 'default'}")
print(f"None or 'fallback': {None or 'fallback'}")
```

```python
# Proving short-circuit: the second operand is never evaluated
def expensive():
    print("  expensive() was called!")
    return 42

print("Test 1: 0 and expensive()")
result = 0 and expensive()       # expensive() NOT called
print(f"  result = {result}")

print("\nTest 2: 1 and expensive()")
result = 1 and expensive()       # expensive() IS called
print(f"  result = {result}")

print("\nTest 3: 'exists' or expensive()")
result = "exists" or expensive() # expensive() NOT called
print(f"  result = {result}")
```

## 2.5 Type Conversion

```python
# Explicit conversion
x = int("42")
y = float(7)
s = str(3.14)
b = bool(0)

print(f"int('42')  = {x} (type: {type(x).__name__})")
print(f"float(7)   = {y} (type: {type(y).__name__})")
print(f"str(3.14)  = {s} (type: {type(s).__name__})")
print(f"bool(0)    = {b} (type: {type(b).__name__})")

# Implicit promotion
result = 10 + 2.5
print(f"\n10 + 2.5 = {result} (type: {type(result).__name__})")
```
