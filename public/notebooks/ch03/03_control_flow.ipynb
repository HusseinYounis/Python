## 3.1 Conditional Statements

````python
temperature = 35

if temperature > 30:
    print("Hot")
elif temperature > 20:
    print("Warm")
elif temperature > 10:
    print("Cool")
else:
    print("Cold")
````

### Truthiness and Falsiness

Falsy values: `False`, `None`, `0`, `0.0`, `""`, `[]`, `{}`, `()`, `set()`  
Everything else is **truthy**.

````python
# Pythonic checks — use truthiness directly
name = ""
items = [1, 2, 3]

if name:
    print(f"Hello, {name}")
else:
    print("Name is empty")

if items:
    print(f"Found {len(items)} items")
````

### Chained Comparisons & Membership Testing

````python
x = 15

# Chained comparison (unique to Python!)
if 10 < x < 20:
    print("x is between 10 and 20")

# Membership testing with `in`
vowels = "aeiou"
char = "e"
if char in vowels:
    print(f"'{char}' is a vowel")
````

### Ternary (Conditional) Expression

````python
# Python: value_if_true if condition else value_if_false
# C++:    (condition) ? value_if_true : value_if_false

age = 20
category = "adult" if age >= 18 else "minor"
print(f"Age {age} → {category}")

score = 85
grade = "A" if score >= 90 else "B" if score >= 80 else "C" if score >= 70 else "F"
print(f"Score {score} → Grade {grade}")
````

## 3.2 Loops

### The `range()` Function

````python
print("range(5):          ", list(range(5)))
print("range(2, 8):       ", list(range(2, 8)))
print("range(0, 20, 3):   ", list(range(0, 20, 3)))
print("range(10, 0, -2):  ", list(range(10, 0, -2)))
````

### `for` Loop

````python
# Iterating over a list
languages = ["Python", "C++", "Java"]
for lang in languages:
    print(lang)

print()

# Iterating over a string
for char in "Python":
    print(char, end=" ")
````

### `enumerate()` — Index + Value

````python
fruits = ["apple", "banana", "cherry"]

# Pythonic way
for i, fruit in enumerate(fruits, start=1):
    print(f"{i}. {fruit}")
````

### `while` Loop

````python
# Countdown
count = 5
while count > 0:
    print(count, end=" ")
    count -= 1
print("Go!")
````

### `break`, `continue`, and Loop `else`

````python
# break — stop at first negative
numbers = [10, 25, -3, 42, 8]
for n in numbers:
    if n < 0:
        print(f"Negative found: {n}")
        break
    print(f"Processing: {n}")
````

````python
# continue — skip odd numbers
for i in range(10):
    if i % 2 != 0:
        continue
    print(i, end=" ")
````

````python
# Loop else — runs only if loop completes WITHOUT break
def is_prime(n):
    if n < 2:
        return False
    for d in range(2, int(n ** 0.5) + 1):
        if n % d == 0:
            return False
    else:
        return True

for num in [2, 7, 15, 17, 20, 23]:
    print(f"{num}: {'prime' if is_prime(num) else 'not prime'}")
````

### Nested Loops

````python
# Multiplication table (5×5)
for i in range(1, 6):
    for j in range(1, 6):
        print(f"{i*j:4d}", end="")
    print()
````

## 3.3 Structural Pattern Matching (`match`/`case`)

*Python 3.10+ — much more powerful than C++/Java `switch`*

````python
# Basic value matching
def describe_http(status):
    match status:
        case 200:
            return "OK"
        case 301:
            return "Moved Permanently"
        case 404:
            return "Not Found"
        case 500:
            return "Server Error"
        case _:
            return f"Unknown: {status}"

for code in [200, 404, 500, 403]:
    print(f"{code}: {describe_http(code)}")
````

````python
# OR patterns and guards
def classify_day(day):
    match day.lower():
        case "saturday" | "sunday":
            return "Weekend"
        case "monday" | "tuesday" | "wednesday" | "thursday" | "friday":
            return "Weekday"
        case _:
            return "Invalid day"

print(classify_day("Saturday"))
print(classify_day("Monday"))
````

````python
# Sequence (destructuring) patterns
def process_command(command):
    match command.split():
        case ["quit"]:
            return "Exiting..."
        case ["hello", name]:
            return f"Hello, {name}!"
        case ["move", direction, distance]:
            return f"Moving {direction} by {distance}"
        case ["add", *numbers]:
            total = sum(int(n) for n in numbers)
            return f"Sum = {total}"
        case _:
            return "Unknown command"

commands = ["quit", "hello Alice", "move north 10", "add 1 2 3 4"]
for cmd in commands:
    print(f"{cmd!r:>20} → {process_command(cmd)}")
````

````python
# Dictionary (mapping) patterns
def process_event(event):
    match event:
        case {"type": "click", "x": x, "y": y}:
            return f"Click at ({x}, {y})"
        case {"type": "keypress", "key": key}:
            return f"Key pressed: {key}"
        case _:
            return "Unknown event"

print(process_event({"type": "click", "x": 100, "y": 200}))
print(process_event({"type": "keypress", "key": "Enter"}))
````

## 3.4 Comprehensions

Concise, readable way to build collections from iterables.

### List Comprehension

````python
# Basic — squares of 0..9
squares = [x ** 2 for x in range(10)]
print("Squares:", squares)

# Filtering — even numbers only
evens = [x for x in range(20) if x % 2 == 0]
print("Evens:", evens)

# Conditional expression
labels = ["even" if x % 2 == 0 else "odd" for x in range(6)]
print("Labels:", labels)
````

### Dictionary and Set Comprehensions

````python
# Dict comprehension — word → length
words = ["hello", "world", "python"]
word_lengths = {w: len(w) for w in words}
print("Word lengths:", word_lengths)

# Set comprehension — unique lengths
words2 = ["cat", "dog", "ant", "elephant", "bee"]
unique_lengths = {len(w) for w in words2}
print("Unique lengths:", unique_lengths)
````

### Nested Comprehensions

````python
# Flatten a 2D matrix
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat = [val for row in matrix for val in row]
print("Flattened:", flat)

# Multiplication table
table = [[i * j for j in range(1, 6)] for i in range(1, 6)]
for row in table:
    print(row)
````

### Generator Expressions

````python
import sys

# Compare memory: list comprehension vs generator expression
list_comp = [x ** 2 for x in range(10_000)]
gen_expr  = (x ** 2 for x in range(10_000))

print(f"List: {sys.getsizeof(list_comp):>8,} bytes")
print(f"Gen:  {sys.getsizeof(gen_expr):>8,} bytes")

# Use generators with any function that accepts an iterable
total = sum(x ** 2 for x in range(1_000_000))
print(f"\nSum of squares (1M): {total:,}")
````

## 3.5 Iteration Utilities

### `zip()` — Parallel Iteration

````python
names = ["Alice", "Bob", "Charlie"]
scores = [95, 87, 92]

for name, score in zip(names, scores):
    print(f"{name}: {score}")

# Create dict from two lists
student_scores = dict(zip(names, scores))
print(f"\nDict: {student_scores}")
````

### `sorted()`, `reversed()`, `any()`, `all()`

````python
numbers = [3, 1, 4, 1, 5, 9, 2, 6]

print("sorted:  ", sorted(numbers))
print("reversed:", list(reversed(sorted(numbers))))

# Sort by custom key
words = ["banana", "apple", "cherry", "date"]
print("by length:", sorted(words, key=len))
````

````python
scores = [85, 92, 78, 45, 91]

print(f"Any > 90?   {any(s > 90 for s in scores)}")
print(f"All >= 60?  {all(s >= 60 for s in scores)}")
````

### `map()` and `filter()`

````python
numbers = [1, 2, 3, 4, 5]

# map — apply function to each element
squared = list(map(lambda x: x ** 2, numbers))
print("Squared:", squared)

# filter — keep matching elements
evens = list(filter(lambda x: x % 2 == 0, numbers))
print("Evens:", evens)

# Convert list of strings to integers
str_nums = ["10", "20", "30"]
int_nums = list(map(int, str_nums))
print("Converted:", int_nums)
````
